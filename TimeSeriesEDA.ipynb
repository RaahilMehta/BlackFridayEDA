```python
df_tesla = pdr.get_data_yahoo('TSLA', start='2017-10-24', end='2022-12-23')
```

```python
df_tesla['High'].plot(figsize=(12,4))
```

```python
##x lijmit and y limit
df_tesla['High'].plot(xlim=['2020-01-01','2021-09-1'],ylim=[0,900],figsize=(12,4))
```

```python
##x lijmit and y limit
df_tesla['High'].plot(xlim=['2020-01-01','2021-09-1'],ylim=[0,900],figsize=(12,4),ls='--',c='green')
```
